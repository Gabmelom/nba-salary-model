```python
import matplotlib.pyplot as plt

# Comparison between models as a bar chart
mse = {
    'Linear Regression': 4.3,
    'Neural Net': 2.9,
    'Random Forest': 3.2
}

r2 = {
    'Linear Regression': 0.65,
    'Neural Net': 0.76,
    'Random Forest': 0.74
}

colors = ['#438cff', '#ff5e5e', '#78b087']

# Make two bar graphs side by side, one for mse and one for r2, grouped by model
fig, ax = plt.subplots(ncols=2, figsize=(10, 6))
ax[0].bar(mse.keys(), mse.values(), color=colors)
ax[0].set_title('Mean Squared Error')
ax[0].set_ylabel('MSE (millions)')
ax[0].grid(axis='y', linestyle='--', alpha=0.4)

ax[1].bar(r2.keys(), r2.values(), color=colors)
ax[1].set_title(r'$R^2$ Score')
ax[1].set_ylabel(r'$R^2$')
ax[1].grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
